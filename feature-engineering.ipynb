```python
import pandas as pd
import numpy as np
```

## Feature Engineering

```python
df = pd.read_csv(
    "cleaned_supplychain_data.csv"
)

df.head()
```

```python
df["order_date"] = pd.to_datetime(
    df["order date (DateOrders)"]
)


df["shipping_date"] = pd.to_datetime(
    df["shipping date (DateOrders)"]
)


df["delivery_time"] = (
    df["shipping_date"] -
    df["order_date"]
).dt.days
```

```python
df["profit"] = df["Benefit per order"]
```

```python
df["inventory_demand"] = (
    df["Order Item Quantity"]
    *
    df["Sales"]
)
```

```python
optimization_data = df[
[
"Delivery Status",
"Shipping Mode",
"delivery_time",
"profit",
"inventory_demand",
"Order Item Quantity"
]
]

optimization_data.head()
```

```python
optimization_data.to_csv(
"optimization_dataset.csv",
index=False
)
```
